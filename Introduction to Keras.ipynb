# Introduction to Keras

DATA PROCESSING

```python
# Import libraries to work with
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

Data Creation

```python
train_samples = []
train_labels = []
```

```python
for i in range (50):
    #The younger peple who had side effects
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    #The older people who had no side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
    
for i in range (1000):
    #The younger peple who had no side effects
    random_younger = randint(13, 64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    #The older people who had side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
```

```python
#Convert train_samples and train_labels into numpy arrays then shuffle them to rearrange the order
train_samples = np.array(train_samples)
train_labels = np.array(train_labels)
train_samples,train_labels = shuffle(train_samples, train_labels)
```

```python
#scale all of the data down from a scale ranging from 13 to 100 to be on a scale from 0 to 1
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# Create an artificial neural network with TensorFlow's Keras API

how to create a simple artificial neural network using a Sequential model from the Keras API integrated within TensorFlow.

```python
#Import libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
#This is the crossentropy metric class to be used when there are multiple label classes (2 or more).
from tensorflow.keras.metrics import categorical_crossentropy 
```

BUILD A SEQUENTIAL MODEL

```python
#A tf.keras.Sequential model is a linear stack of layers. It accepts a list, 
#and each element in the list should be a layer.
model = Sequential([
    Dense(units = 16, input_shape=(1,), activation = 'relu'), #First hidden layer 
    Dense(units= 32, activation = 'relu'),#Second hidden layer
    Dense(units = 2, activation ='softmax') #Output layer
    #the activation function well use is softmax, 
    #which will give us a probability distribution among the possible outputs. 
])
```

```python
model.summary()
```

# Train an artificial neural network with TensorFlow's Keras API

COMPILE THE MODEL

```python
#Metric values are displayed during fit(). Metrics expects a list or a dictionary
#Adam optimization is a stochastic gradient descent (SGD) method
#Adam accepts an optional parameter learning_rate, which well set to 0.0001. 
model.compile(optimizer = Adam(learning_rate=0.0001), loss = 'sparse_categorical_crossentropy', metrics =['accuracy'])
```

# Training and validating the model

```python
#After compiling the model we will train it with the fit() function
#an epoch is a single pass of all the data to the network.
#verbose=2. This just specifies how much output to the console we want to see during each epoch of training.
model.fit(x=scaled_train_samples, y=train_labels, validation_split = 0.1, batch_size=10, epochs=30, verbose=2)
```

Our model is learning the features of the training data, but also how well the model is generalizing to new, unseen data from the validation set. Our model is not overfitting.

# Neural Network Predictions with TensorFlow's Keras API

The hope is that, even though our model wasnt exposed to these particular data during training, it will still be able to accurately make predictions of the data based on what its learned from the training data set from which it was trained.

We call this process inference, as the model is using its knowledge gained from training and using it to infer a prediction or result. 
The test set is the set of data used specifically for inference after training has concluded.

```python
test_samples=[]
test_labels=[]

for i in range(10):
    # the 5% of younger people that had side effects from the use of the drug
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    #the 5% of older people that had no side effects from using the drug 
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)
    
for i in range (200):
    # the 95% of younger people that had no side effects from the use of the drug
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    #the 95% of older people that had side effects from using the drug 
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
    
    
```

```python
#Converting dataset to numpy arrays
test_samples = np.array(test_samples)
test_labels = np.array(test_labels)
#Shuffling dataset
test_samples, test_labels = shuffle(test_samples, test_labels)
#Scaling the data
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
#The output from the predictions won't be relevant for us, so we're setting verbose=0 for no output. 
predictions = model.predict(x=scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions:
    print(i)
```

The first column contains the probability for each patient not experiencing side effects, which is represented by a 0. The second column contains the probability for each patient experiencing side effects, which is represented by a 1. 

```python
# np.argmax returns the indices of the maximum values along an axis.
rounded_predictions = np.argmax(predictions, axis=-1)

for i in rounded_predictions:
    print(i)
```

# Create a Confusion Matrix for Neural Network Predictions

 Although we were able to read the predictions from the model easily, we werent easily able to compare the predictions to the true labels for the test data.

With a confusion matrix, well be able to visually observe how well the model predicts on test data. 

```python
#CONSTRUCTING A CONFUSION MATRIX
#import all the libraries you will use
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import itertools #for efficient looping
%matplotlib inline
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects','had_side_effects']
```

```python
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```
